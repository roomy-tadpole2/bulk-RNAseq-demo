```r
library(tibble)
library(dplyr)
library(ggplot2)
library(DESeq2)
library(biomaRt)
sessionInfo()
```

```r
df <- read.csv("../processed/RNA_seq_unstim_counts.csv", row.names = 1)

max_values <- apply(df, 2, max)

average_values <- apply(df, 2, mean)

# Create an HTML table
html_table <- paste0(
    "<table border='1' style='border-collapse: collapse; width: 50%; text-align: center;'>",
    "<tr><th>Sample</th><th>Max Value</th><th>Average Value</th></tr>",
    paste0(
        "<tr><td>", names(max_values), "</td><td>", max_values, "</td><td>", round(average_values, 2), "</td></tr>",
        collapse = ""
    ),
    "</table>"
)

IRdisplay::display_html(html_table)
```

```r
head(df)
```

```r
sample_info <- data.frame(
    row.names = colnames(df),
    condition = factor(c("qKO", "qKO","qKO","WT", "WT", "WT"))
)
sample_info$condition <- relevel(sample_info$condition, ref = "WT")
```

```r
# show sample info
IRdisplay::display(sample_info)
IRdisplay::display(levels(sample_info$condition))
```

```r
count_matrix <- as.matrix(df)

dds <- DESeqDataSetFromMatrix(countData = count_matrix,
                              colData = sample_info,
                              design = ~ condition)
```

```r
dds <- DESeq(dds)
res <- results(dds)
```

```r
summary(res)
```

```r
res_df <- as.data.frame(res)
# head(res_df,5)
res_df$ensembl_id <- sub("\\..*", "", rownames(res_df))  # remove version suffix
# head(res_df,5)
# Ensembl BioMart
ensembl <- useMart("ensembl", dataset = "mmusculus_gene_ensembl")

mapping <- getBM(
  attributes = c("ensembl_gene_id", "external_gene_name"),
  filters = "ensembl_gene_id",
  values = res_df$ensembl_id,
  mart = ensembl
)

# Merge
res_df <- merge(res_df, mapping, by.x = "ensembl_id", by.y = "ensembl_gene_id", all.x = TRUE)

head(res_df[, c("ensembl_id", "external_gene_name")])

colnames(res_df)[colnames(res_df) == "external_gene_name"] <- "gene_symbol"
```

```r
# the function to search for a specific gene in red_df

search_gene <- function(gene_symbol) {
  result <- res_df %>% filter(gene_symbol == !!gene_symbol)
  if (nrow(result) == 0) {
    return(paste("Gene", gene_symbol, "not found."))
  } else {
    return(result)
  }
}

r <- search_gene("Cdc45")

# display r in a nice format
if (is.character(r)) {
  print(r)
} else {
  print(r[, c("ensembl_id", "gene_symbol", "log2FoldChange", "pvalue", "padj")])
}
```

```r
# sort res_df by p-value ascending
res_df_sorted <- res_df %>% arrange(pvalue)

head(res_df_sorted, 10)
```

```r
ggplot(res_df, aes(x = log2FoldChange)) +
  geom_histogram(bins = 100, fill = "steelblue", color = "white") +
  labs(title = "Distribution of log2FoldChange",
       x = "log2FoldChange",
       y = "Count") +
  theme_minimal()
```
