```python
import matplotlib.pyplot as plt

# Original point
x, y = 0.5, 0.5

# Text offset
ofx, ofy = 0.3, 0.3
text = "text message"

# Final text location
tx, ty = x + ofx, y + ofy

# Plot
fig, ax = plt.subplots()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Plot the text
ax.text(tx, ty, text, fontsize=12, ha='left', va='bottom')

# Now draw a zig-zag line from text back to (x, y)
zigzag_path = [
    (x, y),                    # original point
    (x + ofx * 0.3, y + ofy * 0.3),  # zig
    (x + ofx * 0.6, y + ofy * 0.6),  # zag
    (tx, ty)                   # text point
]

# Unpack and plot
xs, ys = zip(*zigzag_path)
ax.plot(xs, ys, color='black', linewidth=2)

# Optional: mark the original point
ax.plot(x, y, 'o', color='black')

plt.show()
```
